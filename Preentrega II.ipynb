```python
from docx import Document

def read_doc(archivo):
    # Funcion para cargar el texto en un string
    doc = Document(archivo)
    texto_completo = ""
    for parrafo in doc.paragraphs:
        texto_completo += parrafo.text
    return texto_completo

antecedentes = read_doc('Prueba 1 Data Science/antecedentes.docx')
descripcion_puesto = read_doc('Prueba 1 Data Science/descripcion del puesto.docx')

prompt = f'ANTECEDENTES: {antecedentes} PUESTO: {descripcion_puesto}'
prompt
```

### Generación de imagen

```python
def generacion_imagen(prompt, size):
    response = client.images.generate(prompt=prompt, n=1,size = size)
    image_url = response['data'][0]['url']
    return image_url
```
